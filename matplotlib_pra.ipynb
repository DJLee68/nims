```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1,4,9,16])
plt.ylabel('some numbers')
plt.show()
```

```python
x=np.array([1,2,3,4])
y=np.array([1,4,9,16])

plt.plot(x,y,'ro:', x, y/5, 'bs-')
plt.show
```

```python
x=np.linspace(0,2,100)

plt.plot(x,x,label='linear')
plt.plot(x,x**2,label='quadratic')
plt.plot(x,x**3, label='cubic')

plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Simple Plot')
plt.xlim([0,2.2])
plt.ylim([-0,2.10])
plt.legend()

plt.show()
```

```python
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)
```

```python
t1=np.arange(0.0,5.0,0.1)
t2=np.arange(0.0,5.0,0.02)
```

```python
plt.figure(1)
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show
```

```python
np.random.seed(222)
data=np.random.randn(100,2)

plt.scatter(data[:,0], data[:,1])
plt.show()
```

```python
t=np.random.randint(2, size=100)
```

```python
plt.scatter(data[:,0], data[:,1], c=t)
plt.show()
```

```python
plt.scatter(data[t==0,0], data[t==0,1], color='red', marker='o', label='class 0')
plt.scatter(data[t==1,0], data[t==1,1], color='blue', marker='^', label='class 1')

plt.legend()
plt.title('two classes')

plt.savefig('test1.eps')
plt.show()
```
